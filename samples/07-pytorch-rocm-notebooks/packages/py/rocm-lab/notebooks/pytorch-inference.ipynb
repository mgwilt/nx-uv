Backend profile: **ROCm 6.2.4**.

If running in Docker with NVIDIA GPUs, start with `--gpus all` and ensure host GPU prerequisites are installed.


```python
import torch

EXPECTED_BACKEND = "rocm"
print(f"PyTorch version: {torch.__version__}")
print(f"Built for CUDA: {torch.version.cuda}")
print(f"Built for ROCm/HIP: {torch.version.hip}")
print(f"CUDA visible at runtime: {torch.cuda.is_available()}")

device = "cuda" if torch.cuda.is_available() else "cpu"
x = torch.randn(1, 4, device=device)
model = torch.nn.Sequential(
    torch.nn.Linear(4, 8),
    torch.nn.ReLU(),
    torch.nn.Linear(8, 2),
).to(device)
with torch.no_grad():
    y = model(x)

print(f"Device used: {device}")
print(f"Output tensor: {y}")
print(f"Expected backend profile: {EXPECTED_BACKEND}")
```
